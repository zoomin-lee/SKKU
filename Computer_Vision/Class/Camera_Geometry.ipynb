# Chap 2. Geometric Camera Models

## Factors in image formation
1️⃣ Geometry

: Relationship between points in the 3D world and their images

2️⃣ Radiometry

: Relationship between the amount of light radiating from a surface and the amount of incident at its image
- Measurement of optical radiation (ultraviolet, visible and infrared)

3️⃣ Photometry

: Ways of measuring the intensity of light
- Measurement of light that is detectable by human eye

4️⃣ Digitization

: Ways of converting continuous signals to digital approximations


<br>

## Image Transformations
1️⃣ Image Filtering : Change range of image ( Intensity )


2️⃣ Image Warping : Change domain of image

<br>

## 👉 2D Linear Transformations

: Scale, Rotation, Shear, Mirror

- Translation is not a linear operation on 2D coordinates❗️

🧐 Properties of Linear Transformations

- origin maps to origin
- lines map to lines
- parallel lines remain parallel
- ratios are preserved
- closed under composition

## 👉 Affine Transformation
: Linear Transformation + Translation

🧐 Properties of Affine Transformation
- **Origin does not necessarily map to origin**
- Lines map to lines
- Parallel lines remain parallel
- Ratios are preserved
- Closed under composition

<br>

## 👉 Perspective(= Projective ) Transformation = Homography
: Affine Transformation + Projective warps

🧐 Properties of Perspective Transformation
- **Origin does not necessarily map to origin**
- Lines map to lines
- **Parallel lines do not necessariliy remain parallel**
- **Ratios are not preserved**
- Closed under composition

## Projection Properties
1️⃣ Many to One : any points along same ray map to same point in image

2️⃣ Points -> Points

3️⃣ Lines -> Lines ( collinearity is preserved )
- BUT, line through focal point projects to a point

4️⃣ Planes -> Planes ( or half-planes )
- BUT, plane through focal point projects to line

5️⃣ Parallel lines converge at a vanishing point
- BUT, parallels parallel to the image plane remain parallel

<br>

- Pinhole too big : Many directions are averaged ⏩ Blurring the image
- Pinhole too small : Diffraction effects blur the image
- Pinhole cameras are dark, because very small set of rays from a particular point hits the screen

## Cameras with lenses


image.png

- Rays passing through the center are not deviated (pinhole projection model still holds)
- A lens focuses parallel rays onto a single focal point 
- Focal length is a function of the shape and index of refraction of the lens
- Aperture restricts the range of rays 
  - Aperture may be on either side of the lens
