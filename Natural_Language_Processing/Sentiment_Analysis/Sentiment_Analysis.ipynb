```python
! pip install contractions
import re
import contractions
```

```python
def text_clean(text):
    
    text = contractions.fix(text) #he's -> he is
    text = re.sub('<br />'," ",text)  
    text = re.sub('https?:/\/\S+', ' ', text) # remove urls
    text = re.sub('[0-9]+', ' ', text) # remove numbers
    text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\…》]+', ' ', text) # remove all symbols and punctuation except for . , ! and ?
    text = re.sub("'"," ", text)
    text = re.sub('\s+', ' ', text) # 중복 띄어쓰기 삭제

    text = text.lower() # 소문자
    return text.strip() # 앞 뒤 문장 공백 제거
```

```python
# import nltk
# nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

analyser = SentimentIntensityAnalyzer()

def vader_polarity(text):
  """ Transform the output to a binary 0/1 result """
  score = analyser.polarity_scores(text)
  return 1 if score['pos'] > score['neg'] else 0

for index, (text, sentiment) in enumerate(zip(imdb_df.iloc[:4].body, imdb_df.iloc[:4].sentiment)):

  print(f'Text {index} : {text.strip()}')
  print('Sentiment:', sentiment)
  print('Predicted Sentiment polarity:', analyser.polarity_scores(text))
  print('Predicted Sentiment polarity Class:', vader_polarity(text))
  print('-'*60)
```

```python
# import nltk
# nltk.download('wordnet')
# nltk.download('sentiwordnet')
    
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet as wn
from nltk.corpus import sentiwordnet as swn
from nltk import sent_tokenize, word_tokenize, pos_tag
 
lemmatizer = WordNetLemmatizer()
 
def penn_to_wn(tag):
    """
    Convert between the PennTreebank tags to simple Wordnet tags
    """
    if tag.startswith('J'):
        return wn.ADJ
    elif tag.startswith('N'):
        return wn.NOUN
    elif tag.startswith('R'):
        return wn.ADV
    elif tag.startswith('V'):
        return wn.VERB
    return None
 
def swn_polarity(text):
    """
    Return a sentiment polarity: 0 = negative, 1 = positive
    """
    sentiment = 0.0
    tokens_count = 0
 
    raw_sentences = sent_tokenize(text)
    for raw_sentence in raw_sentences:
        tagged_sentence = pos_tag(word_tokenize(raw_sentence)) # pos_tag : 어떤 형태소인지 뽑아주는 nltk 함수
 
        for word, tag in tagged_sentence:
            wn_tag = penn_to_wn(tag)
            if wn_tag not in (wn.NOUN, wn.ADJ, wn.ADV):
                continue
 
            lemma = lemmatizer.lemmatize(word, pos=wn_tag)
            if not lemma:
                continue
 
            synsets = wn.synsets(lemma, pos=wn_tag)
            if not synsets:
                continue
 
            # Take the first sense, the most common = 동음이의어 중 가장 첫번째 뜻을 가져옴
            synset = synsets[0]
            swn_synset = swn.senti_synset(synset.name())
 
            sentiment += swn_synset.pos_score() - swn_synset.neg_score() # 문장의 긍/부정을 따져야 하므로, 단어당 가진 긍/부정 score를 합쳐줌
            tokens_count += 1
    print(synset.name())
    # judgment call ? Default to positive or negative
    if not tokens_count:
        return 0
 
    # sum greater than 0 => positive sentiment
    if sentiment >= 0:
        return sentiment, 1
 
    # negative sentiment
    return sentiment, 0
```

```python
 for index, (text, sentiment) in enumerate(zip(imdb_df.iloc[:4].body, imdb_df.iloc[:4].sentiment)):
    print(f'Text {index} : {text.strip()}')
    print('Sentiment:', sentiment)
    print('Predicted Sentiment polarity:', swn_polarity(text))
    print('-'*60)
```

```python
import re
from soynlp.normalizer import repeat_normalize
from soynlp.normalizer import only_hangle_number
from hanspell import spell_checker

def ko_text_clean(text):
    
    text = only_hangle_number(text) # 한글만 남김 ex, 특수문자나 이런것들 지워줌
    text = repeat_normalize(text, num_repeats=2) # 2번 이상 반복적으로 쓴 것 지워줌
    spell = spell_checker.check(text) # 맞춤법과 띄어쓰기 고쳐줌
    text = spell.checked
    text = re.sub('\s+', ' ', text) # 중복 띄어쓰기 삭제

    return text.strip()
```

```python
import pandas as pd
from konlpy.tag import Komoran
komoran = Komoran()
```

```python
lala_review = pd.read_csv('/content/drive/MyDrive/Data_Mining/DM_Assignment5/dataset/naver_preprocess.csv',encoding='utf-8')
lala_review = lala_review.dropna() #전처리 후에 빈칸으로만 남아있는 데이터 줄 삭제
```

```python
polarity_all = pd.read_csv('/content/drive/MyDrive/Data_Mining/DM_Assignment5/dataset/polarity.csv',encoding='utf-8')
polarity_all
```

```python
lala_review['sentiment'] = [1 if lala_review['score'].iloc[i] >= 6 else 0 for i in range(len(lala_review))]
lala_review.head(10)
```

```python
polarity_del = polarity_all[polarity_all['ngram'].str.contains(';')].index
polarity_del
```

```python
polarity = polarity_all.drop(index = polarity_del)
polarity.head(10)
```

```python
pol={}
for n in range(len(polarity)):
    ngram = polarity['ngram'].iloc[n]
    pos = polarity['POS'].iloc[n]
    neg = polarity['NEG'].iloc[n]
    ngram = ngram.split('/')
    ngram[0]= ngram[0].replace("*","")
    pol[(ngram[0],ngram[1])] = (pos,neg)
pol
```

```python
def analyzer(text):
  sentence = komoran.pos(text)
  sentiment = 0
  pscore = 0
  nscore = 0
  for group in sentence :
    if group in pol :
      pscore += pol[group][0]
      nscore += pol[group][1]
  score = pscore - nscore
  if score >= 0:
    sentiment = 1
  else :
    sentiment = 0
  return sentiment, score
```

```python
correct = 0
pred_sentiment = []
for i in range(len(lala_review)):
  pred_sent, pred_score = analyzer(lala_review['body'].iloc[i])
  pred_sentiment.append(pred_sent)
  label = lala_review['sentiment'].iloc[i]
  if label == pred_sent :
    correct += 1
  print('index: ', i)
  print('Sentiment: ', lala_review['sentiment'].iloc[i])
  print('Predicted Sentiment polarity: ', pred_score)
  print('Predicted Sentiment polarity Class: ', pred_sent)
  print('-'*60)
accuracy = round((correct / len(lala_review) * 100), 2)
print('Accuracy : ', accuracy)
```
