- 차원이 매우 크다면 차원의 저주에 의해 어떤 알고리즘보다 Brute Force 전략이 더 효율적임

## Recursion vs Iterative 

### 👉 피보나치 수열

1️⃣ Recursion : 동일한 피보나치 수를 중복하여 계산하므로 비효율적임 

- Divide and Conquer 전략을 통해 설계된 알고리즘


T(n) : fibo(n)을 계산하기 위하여 fibo() 함수를 호출하는 횟수

    T(0) = 1
    T(1) = 1
    T(n) = T(n-1) + T(n-2) 
         > 2 x T(n-2)
         > 2 x 2 x T(n-4)
         ...
         > 2 x 2 x ... x 2 x T(0) 
         = 2^(n/2) 

```python
def fibo(n) :
  if n == 0 or n == 1 :
    return n
  else :
    return fibo(n-1) + fibo(n-2)
```

2️⃣ Iterative : 중복 계산이 없으므로 효율적임

- Dynamic Programming 전략을 통해 설계된 알고리즘

```python
def fibo(n) :
  memo = [0 for _ in range(n+1)]
  memo[1] = 1
  for i in range(2, n+1):
    memo[i] = memo[i-1] + memo[i-2]
  return memo[n]
```

### 👉 Factorial 계산
**Iterative Alogrithm이 Recursive Algorihm보다 항상 효율적인가?**

: 대부분의 경우 Iterative가 Recursive보다 효율적임
- 하지만, 알고리즘 설계 단계에서 재귀 알고리즘은 유용하며, 때로는 재귀적으로 설계해도 매우 효율적일 수 있음
- 재귀 호출 횟수와 반복의 횟수가 동일하면 Iterative Alogrithm이 상수적으로만 효율적임

1️⃣ Recursion

T(n) : factorial(n)을 계산하기 위하여 factorial() 함수를 호출하는 횟수

    T(0) = 1
    T(n) = 1 + T(n-1)
         = 2 x T(n-2)
         ...
         = n-1 x T(1)
         = n

```python
def factorial(n) : 
  if n <= 1 :
    return 1
  else :
    return n * factorial(n-1)
```

2️⃣ Iterative

```python
def factorial(n) :
  result = 1
  for i in range(1, n+1) :
    result *= i
  return result
```

## Time Complexity 분석
: **입력 크기 n**에 따라서 **단위 연산**(비교연산, 할당연산 등)이 몇 번 수행되는지 결정하는 절차

1️⃣ 최악의 경우 W(n) 

: ( 입력크기 + 입력값 )에 대해 기본연산 수행 횟수가 최대인 경우

2️⃣ 최선의 경우 B(n) 

: ( 입력크기 + 입력값 )에 대해 기본연산 수행 횟수가 최소인 경우


3️⃣ 평균의 경우 A(n) 

: ( 입력크기+입력값 )에 대해 입력값의 확률분포를 가정 후, 모든 입력값에 대해 기본연산이 수행되는 횟수의 평균

Linear Search의 경우 

    x가 L 내에 존재할 확률 = p
    x가 L 내에 i 번째 있을 확률 = p/n
    x가 L 내에 존재하지 않을 확률 = 1-p
    따라서, A(n) = sum( i x p/n ) + n(1-p)
                 = (n+1)p/2 + n(1-p)
                 = n(1-p/2) + p/2

4️⃣ 모든 경우 T(n) 

: 입력값의 내용과 무관하게 복잡도가 항상 일정한 경우에만 가능 

= 시간복잡도는 입력 크기에만 종속적임
- 예를 들어, Linear Search Algorithm은 리스트 항목 값에 따라서 검색하는 횟수가 달라지므로 모든 경우 분석이 불가능함
- 반면에, 리스트 내의 모든 항목값을 더하는 연산의 경우 입력 크기에만 종속적이므로 모든 경우 분석 가능함

- W(n) = B(n) = A(n)인 경우 모든 경우의 분석 사용

## Asymptotic Notation (점근적 표기법)
: 입력 크기 n의 값이 무한히 큰 경우일 때 알고리즘의 복잡도

image.png

✔️ O(n!) < O(2^n) < O(n^3) < O(n^2) < O(nlogn) < O(n) < O(logn) < O(1)

- f(n) ∈ O(g(n)) ➡️ g(n) ∈ Ω(f(n))
- f(n) ∈ Θ(g(n)) ➡️ g(n) ∈ Θ(f(n))
- b > 1이고 a > 1이면, log_a n ∈ Θ(log_b n)은 항상 성립
  - log 복잡도 함수는 모두 같은 카테고리에 속함
- 지수 복잡도는 모두 같은 카테고리에 속하지 않음
  - b > a > 0이면, a^n ∈ o(b^n)


1️⃣ O( Big-Oh ) 표기법

: 알고리즘 복잡도의 점근적 상한을 표시 

    n > n0인 모든 정수 n에 대해서 f(n) <= c * g(n)이 성립하는  c > 0, 음이 아닌 정수 n0가 존재하면, f(n) ∈ O(g(n))임

    f(n)은 g(n)의 Big-Oh에 속함

🧐 n^3 ∈ O(n^2)? ❌

: 양변을 n^3으로 나누면 n <= c가 되는데, c에 아무리 큰 수를 대입해도 그 보다 더 큰 n이 존재하므로 성립하지 않음

<br>

2️⃣ Ω( Omega ) 표기법

: 알고리즘 복잡도의 점근적 하한을 표시 

    n > n0인 모든 정수 n에 대해서 f(n) >= c * g(n)이 성립하는 c > 0, 음이 아닌 정수 n0가 존재하면, f(n) ∈ Ω(g(n))임

    f(n)은 g(n)의 Omega에 속함

🧐 n ∈ O(n^2)? ❌

: 양변을 cn으로 나누면 1/c >= n이 되는데 1/c보다 큰 n항은 항상 존재하므로 성립하지 않음

<br>

🌟 Optimal한 알고리즘 

: 어떤 문제가 Ω(f(n))의 시간복잡도를 가진다고 알려져있을 때, 그 문제를 O(f(n))만에 푸는 알고리즘을 optimal하다고 합니다.

<br>

3️⃣ Θ( Theta ) 표기법

: Θ(g(n)) = O(g(n)) ∩ Ω(g(n))
  
    n > n0인 모든 정수 n에 대해서 c1 * g(n) <= f(n) <= c2 * g(n)이 성립하는 c1 > 0, c2 > 0, 음이 아닌 정수 n0이 존재하면 f(n) ∈ Θ(g(n))

    f(n)은 g(n)과 동일한 차수(order)임

- Θ(g(n))은 복잡도 카테고리를 나눌 때 사용

<br>

4️⃣ o( Small-Oh ) 표기법

    n > n0인 모든 정수 n과 모든 양의 실수 c에 대해서 항상 f(n) <= c * g(n)이 성립하면, f(n) ∈ o(g(n))임

    f(n)은 g(n)의 Small-Oh에 속함

- Big-Oh는 f(n) <= c * g(n)을 만족하는 **어떤 양의 실수 c**가 존재
- Small-Oh는 f(n) <= c * g(n)을 만족하는 **모든 양의 실수 c**가 존재
