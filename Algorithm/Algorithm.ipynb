## Recursion vs Iterative 

### üëâ ÌîºÎ≥¥ÎÇòÏπò ÏàòÏó¥

1Ô∏è‚É£ Recursion : ÎèôÏùºÌïú ÌîºÎ≥¥ÎÇòÏπò ÏàòÎ•º Ï§ëÎ≥µÌïòÏó¨ Í≥ÑÏÇ∞ÌïòÎØÄÎ°ú ÎπÑÌö®Ïú®Ï†ÅÏûÑ 

- Divide and Conquer Ï†ÑÎûµÏùÑ ÌÜµÌï¥ ÏÑ§Í≥ÑÎêú ÏïåÍ≥†Î¶¨Ï¶ò


T(n) : fibo(n)ÏùÑ Í≥ÑÏÇ∞ÌïòÍ∏∞ ÏúÑÌïòÏó¨ fibo() Ìï®ÏàòÎ•º Ìò∏Ï∂úÌïòÎäî ÌöüÏàò

    T(0) = 1
    T(1) = 1
    T(n) = T(n-1) + T(n-2) 
         > 2 x T(n-2)
         > 2 x 2 x T(n-4)
         ...
         > 2 x 2 x ... x 2 x T(0) 
         = 2^(n/2) 

```python
def fibo(n) :
  if n == 0 or n == 1 :
    return n
  else :
    return fibo(n-1) + fibo(n-2)
```

2Ô∏è‚É£ Iterative : Ï§ëÎ≥µ Í≥ÑÏÇ∞Ïù¥ ÏóÜÏúºÎØÄÎ°ú Ìö®Ïú®Ï†ÅÏûÑ

- Dynamic Programming Ï†ÑÎûµÏùÑ ÌÜµÌï¥ ÏÑ§Í≥ÑÎêú ÏïåÍ≥†Î¶¨Ï¶ò

```python
def fibo(n) :
  memo = [0 for _ in range(n+1)]
  memo[1] = 1
  for i in range(2, n+1):
    memo[i] = memo[i-1] + memo[i-2]
  return memo[n]
```

### üëâ Factorial Í≥ÑÏÇ∞
**Iterative AlogrithmÏù¥ Recursive AlgorihmÎ≥¥Îã§ Ìï≠ÏÉÅ Ìö®Ïú®Ï†ÅÏù∏Í∞Ä?**

: ÎåÄÎ∂ÄÎ∂ÑÏùò Í≤ΩÏö∞ IterativeÍ∞Ä RecursiveÎ≥¥Îã§ Ìö®Ïú®Ï†ÅÏûÑ
- ÌïòÏßÄÎßå, ÏïåÍ≥†Î¶¨Ï¶ò ÏÑ§Í≥Ñ Îã®Í≥ÑÏóêÏÑú Ïû¨Í∑Ä ÏïåÍ≥†Î¶¨Ï¶òÏùÄ Ïú†Ïö©ÌïòÎ©∞, ÎïåÎ°úÎäî Ïû¨Í∑ÄÏ†ÅÏúºÎ°ú ÏÑ§Í≥ÑÌï¥ÎèÑ Îß§Ïö∞ Ìö®Ïú®Ï†ÅÏùº Ïàò ÏûàÏùå
- Ïû¨Í∑Ä Ìò∏Ï∂ú ÌöüÏàòÏôÄ Î∞òÎ≥µÏùò ÌöüÏàòÍ∞Ä ÎèôÏùºÌïòÎ©¥ Iterative AlogrithmÏù¥ ÏÉÅÏàòÏ†ÅÏúºÎ°úÎßå Ìö®Ïú®Ï†ÅÏûÑ

1Ô∏è‚É£ Recursion

T(n) : factorial(n)ÏùÑ Í≥ÑÏÇ∞ÌïòÍ∏∞ ÏúÑÌïòÏó¨ factorial() Ìï®ÏàòÎ•º Ìò∏Ï∂úÌïòÎäî ÌöüÏàò

    T(0) = 1
    T(n) = 1 + T(n-1)
         = 2 x T(n-2)
         ...
         = n-1 x T(1)
         = n

```python
def factorial(n) : 
  if n <= 1 :
    return 1
  else :
    return n * factorial(n-1)
```

2Ô∏è‚É£ Iterative

```python
def factorial(n) :
  result = 1
  for i in range(1, n+1) :
    result *= i
  return result
```
